```python
#-----------------------------------------recria a sequencia de Fibonacci-----------------------------------------------
# eu defini para que o codigo realizasse a execução 20 vezes, mas caso seja necessário mais, basta alterar o numero no while
def Fibonacci():
  indice = 0
  soma = 1
  lista = [0,1]

  while indice < 19:

      soma = soma + lista[indice]
      lista.append(soma)

      indice = indice + 1

  return lista;

#-------------------------------Identifica se um item existe na lista--------------------------------
def numlist(n, l):
  x = n in l
  return x


#----------Recebe o numero do usuário e executa as funções, também aplica um tratamento de erros, caso o usuário digite um caractere que não seja um numero inteiro--------------
try:

  num = int(input('Digite um numero: '))

  lista = Fibonacci()
  existe = numlist(num, lista)

  if existe == True:
      print(f':) O numero {num} existe dentro da sequencia de Fibonacci!!!!!!!!')
  else:
      print(f'Infelizmente o numero {num} não existe dentro da sequencia de Fibonacci')

except:
  print('reinicie o programa e adicione um valor inteiro')



```
